## Boxplot
used to find outliers
data.describe() - 25% 50% 75%
box plot has same 3 quarters 
below the box, 25%, 
box - 50%
above the box - 25%
orange line - meadian point - central representation of data

quartiles - arrange data in quarters
interquartile - way of calculatinf range of data

min value - lowermost horizontal line
max value - topmost horizontal line

small circles - outliers:- points lying outside the 1st and 3rd quartile (points which lie outside the group, do not belong in that specific group)

```python
data1=[np.random.normal(20,std,165) for std in range(1,4)]
data1
```

```python
plt.boxplot(data1)
plt.show()
```

```python
import math

#to see the data
x=np.arange(0,math.pi*2,0.05)    #step of 0.5
x
y=np.sin(x)
y
```

```python
# creating blank of empty canvas
fig1=plt.figure()

# adding the axes
ax=fig1.add_axes([0,0,0.5,0.5])     #to plot in the entire canvas. half canvas - 0,0,0.5,0.5
#shrinking of plot from 1 to 0.5
# label the axes
ax.set_title('Sine Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.plot(x,y)
plt.show()
```

```python
# common layouts plots

fig,a=plt.subplots(2,2)
x=np.arange(1,6)
a[0][0].plot(x,x*x) #squaring
a[0][0].set_title('First Plot')
a[0][1].hist(x)
a[0][1].set_title('Histogram')
a[1][0].plot(x,np.sqrt(x))
a[1][0].set_title('Square Root')
a[1][1].plot(x,np.log(x))
a[1][1].set_title('Log')
plt.show()
```

```python
x=np.arange(0,math.pi*2)  #here, pi=3.14 ---> need to convert into degree as graph shows radian
y=np.tan(x)
```

```python
# creating blank of empty canvas
fig1=plt.figure()

# adding the axes
ax=fig1.add_axes([0,0,1,1])     #to plot in the entire canvas. half canvas - 0,0,0.5,0.5
#shrinking of plot from 1 to 0.5
# label the axes
ax.set_title('Tan Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Tangent')
ax.plot(x,y)
plt.show()
```
